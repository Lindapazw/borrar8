# Visualizacion N°1

Este gráfico muestra la cantidad de autos Honda vendidos según su modelo. Es importante ya que proporciona información valiosa para la marca sobre cuáles son los modelos más populares y demandados por los consumidores, lo que puede ayudar a la empresa a tomar decisiones estratégicas para mejorar sus ventas y aumentar su participación en el mercado. Además, también puede ser útil para los vendedores de autos Honda para saber cuáles son los modelos más buscados por los compradores y poder adaptar su enfoque de ventas en consecuencia.


```python
# Obtener el recuento de lo modelos de autos y ordenar en orden descendente
modelos = df['Model'].value_counts().sort_values(ascending=False)

# Crear el gráfico de barras
plt.figure(figsize=(27, 6))  # Tamaño del gráfico (opcional)
sns.countplot(x='Model', data=df, order=modelos.index)  # Ordenar en orden descendente
plt.xticks(rotation=90)
plt.xlabel('Modelos')
plt.ylabel('Cantidad')
plt.title('Cantidad de autos según su modelo')

# Mostrar el gráfico
plt.show()
```

# Visualizacion N°2

```python
# Obtener el recuento de las condiciones de los autos Honda
condiciones = df['Condition'].value_counts()

# Crear el gráfico de pie
plt.pie(condiciones, labels=condiciones.index, autopct='%1.1f%%')
plt.title('Condiciones de los autos Honda')

# Agregar el recuadro de significado de los colores
plt.legend(title="Significado", labels=["Nuevo", "Usado", "Certificado"], loc="center left", bbox_to_anchor=(1.3, 0.5))

# Mostrar el gráfico
plt.show()
```
